```python
import speech_recognition as sr

# Define a list of specific words and corresponding audio file paths
target_words_and_files = {
    "Apple": "/content/drive/MyDrive/wav_files/apple.wav",
    # "chair": "/content/drive/MyDrive/wav_files/chair1.wav",
    # "stair": "/content/drive/MyDrive/wav_files/stair1.wav",
    "table": "/content/drive/MyDrive/wav_files/cable.wav",
    "laptop": "/content/drive/MyDrive/wav_files/shaptop.wav",
    "Charger": "/content/drive/MyDrive/wav_files/sharger.wav",
    "mouse": "/content/drive/MyDrive/wav_files/mouse.wav",
    "Rout": "/content/drive/MyDrive/wav_files/shout.wav",
    "Razzer": "/content/drive/MyDrive/wav_files/razzer.wav"
}

# Initialize the SpeechRecognition recognizer
recognizer = sr.Recognizer()

# Loop through each target word and corresponding audio file
correct_count = 0
wrong_count = 0

for target_word, file_path in target_words_and_files.items():
    print(f"Please speak: {target_word}")

    try:
        # Transcribe the audio from the file
        with sr.AudioFile(file_path) as source:
            audio = recognizer.record(source)

        transcribed_text = recognizer.recognize_google(audio)
        print(f"Transcribed: {transcribed_text}")

        # Check if the transcribed text matches the target word
        if transcribed_text.lower() == target_word.lower():
            print("Correct!")
            correct_count += 1
        else:
            print("Incorrect!")
            wrong_count += 1

    except sr.UnknownValueError:
        print("Speech Recognition could not understand audio.")
        wrong_count += 1

    except sr.RequestError as e:
        print(f"Could not request results from Google Speech Recognition service; {e}")
        wrong_count += 1

# Print the results
print(f"\nResults: Correct - {correct_count}, Incorrect - {wrong_count}")
```

```python
import torch
from transformers import Wav2Vec2ForCTC, Wav2Vec2Processor
import sounddevice as sd

# Load the pretrained Wav2Vec2 model and processor
processor = Wav2Vec2Processor.from_pretrained("facebook/wav2vec2-base-960h")
model = Wav2Vec2ForCTC.from_pretrained("facebook/wav2vec2-base-960h")

# Create the word list
word_list = ["apple", "banana", "cherry", "dog", "cat", "fish", "house", "tree", "sky", "sun"]

# Variables to store results
correct_count = 0
incorrect_count = 0

# Iterate through the word list
for word in word_list:
    print("Speak the word:", word)

    # Record audio
    audio, sample_rate = sd.rec(int(model.sample_rate * model.get_input_duration(processor)), samplerate=model.sample_rate, channels=1)
    sd.wait()

    # Preprocess audio
    input_values = processor(audio, return_tensors="pt").input_values

    # Run transcription without beam search (no correction)
    with torch.no_grad():
        logits = model(input_values).logits
        predicted_ids = torch.argmax(logits, dim=-1)
        transcription = processor.decode(predicted_ids[0])

    # Check for correct match
    if transcription.lower() == word.lower():
        print("Correct!")
        correct_count += 1
    else:
        print("Incorrect. You said:", transcription)
        incorrect_count += 1

# Display final scores
print("Total correct:", correct_count)
print("Total incorrect:", incorrect_count)
```

```python
import librosa
import wavio
import matplotlib.pyplot as plt
import numpy as np

file_path = '/content/download.wav'
y, sr = librosa.load(file_path)

def mfcc_manipulation(audio, n_mfcc=13):
    # Extract MFCC features
    mfccs = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=n_mfcc)

    # Manipulate MFCC features (you can perform any desired modification here)
    # For example, doubling the values of the MFCCs
    mfccs_modified = mfccs * 1

    # Reconstruct audio from modified MFCCs
    y_reconstructed = librosa.feature.inverse.mfcc_to_audio(mfccs_modified)

    return y_reconstructed

# Example of MFCC Manipulation
y_mfcc_manipulated = mfcc_manipulation(y)

if y_mfcc_manipulated is not None:
    output_path_mfcc = '/content/output/output_mfcc_manipulated.wav'

    wavio.write(output_path_mfcc, y_mfcc_manipulated, sr, sampwidth=3)

    # Plot the original audio
    plt.figure(figsize=(12, 4))
    plt.subplot(2, 1, 1)
    plt.plot(np.arange(len(y)) / sr, y)
    plt.title('Original Audio')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')

    # Plot the audio with MFCC manipulation
    plt.subplot(2, 1, 2)
    plt.plot(np.arange(len(y_mfcc_manipulated)) / sr, y_mfcc_manipulated)
    plt.title('Audio with MFCC Manipulation')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')

    plt.tight_layout()
    plt.show()

else:
    print("Error: y_mfcc_manipulated is None. Please check the MFCC manipulation operation.")
```

```python
import librosa
import wavio
import matplotlib.pyplot as plt
import numpy as np

file_path = '/content/download.wav'
y, sr = librosa.load(file_path)

def chroma_manipulation(audio):
    # Extract chroma features
    chroma = librosa.feature.chroma_stft(y=audio, sr=sr)

    # Manipulate chroma features (e.g., transpose)
    chroma_modified = np.transpose(chroma)

    # Reconstruct audio using Griffin-Lim
    y_reconstructed = librosa.griffinlim(chroma_modified)

    return y_reconstructed

# Example of Chroma Feature Manipulation
y_chroma_manipulated = chroma_manipulation(y)

if y_chroma_manipulated is not None:
    output_path_chroma = '/content/output/output_chroma_manipulated.wav'

    wavio.write(output_path_chroma, y_chroma_manipulated, sr, sampwidth=3)

    # Plot the original audio
    plt.figure(figsize=(12, 4))
    plt.subplot(2, 1, 1)
    plt.plot(np.arange(len(y)) / sr, y)
    plt.title('Original Audio')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')

    # Plot the audio with Chroma feature manipulation
    plt.subplot(2, 1, 2)
    plt.plot(np.arange(len(y_chroma_manipulated)) / sr, y_chroma_manipulated)
    plt.title('Audio with Chroma Feature Manipulation')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')

    plt.tight_layout()
    plt.show()

else:
    print("Error: y_chroma_manipulated is None. Please check the Chroma feature manipulation operation.")
```

```python
#Filtered Low Pass in voice

import librosa
import wavio
import matplotlib.pyplot as plt
import numpy as np

file_path = '/content/download.wav'
y, sr = librosa.load(file_path)

def apply_filter(audio, filter_type='low_pass', cutoff_frequency=1000):
    # Apply low-pass filter
    if filter_type == 'low_pass':
        filtered_audio = librosa.effects.preemphasis(audio, coef=cutoff_frequency)
    # Apply high-pass filter
    elif filter_type == 'high_pass':
        filtered_audio = audio - librosa.effects.preemphasis(audio, coef=cutoff_frequency)
    else:
        print("Invalid filter type. Please use 'low_pass' or 'high_pass'.")
        return None

    return filtered_audio

# Example of applying a low-pass filter
filter_type = 'low_pass'  # Change to 'high_pass' for high-pass filter
cutoff_frequency = 0.97  # Adjust as needed
y_filtered = apply_filter(y, filter_type=filter_type, cutoff_frequency=cutoff_frequency)

if y_filtered is not None:
    output_path_filtered = f'/content/output/output_filtered_{filter_type}.wav'

    wavio.write(output_path_filtered, y_filtered, sr, sampwidth=3)

    # Plot the original audio
    plt.figure(figsize=(12, 4))
    plt.subplot(2, 1, 1)
    plt.plot(np.arange(len(y)) / sr, y)
    plt.title('Original Audio')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')

    # Plot the filtered audio
    plt.subplot(2, 1, 2)
    plt.plot(np.arange(len(y_filtered)) / sr, y_filtered)
    plt.title(f'Audio with {filter_type.capitalize()} Filter')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')

    plt.tight_layout()
    plt.show()

else:
    print(f"Error: y_filtered is None. Please check the {filter_type} filter operation.")
```

```python
import librosa
import wavio
import matplotlib.pyplot as plt
import numpy as np

file_path = '/content/download.wav'
y, sr = librosa.load(file_path)

def apply_modulation_effect(audio, effect_type='chorus', modulation_rate=1.0, modulation_depth=0.5):
    # Simulating chorus, flanger, or phaser effects by applying pre-emphasis
    # Adjust the modulation_rate and modulation_depth as needed
    if effect_type == 'chorus':
        modulation_function = np.sin(2 * np.pi * modulation_rate * np.arange(len(audio)) / sr)
    elif effect_type == 'flanger':
        modulation_function = np.sin(2 * np.pi * modulation_rate * np.arange(len(audio)) / sr)
    elif effect_type == 'phaser':
        modulation_function = np.sin(2 * np.pi * modulation_rate * np.arange(len(audio)) / sr)
    else:
        raise ValueError("Invalid effect_type. Choose 'chorus', 'flanger', or 'phaser'.")

    # Apply modulation effect
    audio_modulated = audio + modulation_depth * modulation_function

    return audio_modulated

# Example of applying modulation effect
modulation_effect_type = 'phaser'  # Change to 'flanger' or 'phaser' for different effects
modulation_rate = 1.0  # Adjust as needed
modulation_depth = 0.5  # Adjust as needed
y_modulated = apply_modulation_effect(y, effect_type=modulation_effect_type,
                                       modulation_rate=modulation_rate, modulation_depth=modulation_depth)

if y_modulated is not None:
    output_path_modulated = '/content/output/output_modulated.wav'

    wavio.write(output_path_modulated, y_modulated, sr, sampwidth=3)

    # Plot the original audio
    plt.figure(figsize=(12, 4))
    plt.subplot(2, 1, 1)
    plt.plot(np.arange(len(y)) / sr, y)
    plt.title('Original Audio')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')

    # Plot the modulated audio
    plt.subplot(2, 1, 2)
    plt.plot(np.arange(len(y_modulated)) / sr, y_modulated)
    plt.title(f'Audio with {modulation_effect_type.capitalize()} Effect')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')

    plt.tight_layout()
    plt.show()

else:
    print("Error: y_modulated is None. Please check the modulation effect operation.")
```

```python
import librosa
import numpy as np
import wavio
import matplotlib.pyplot as plt
from scipy.signal import butter, lfilter

# Load the original audio
file_path = '/content/download.wav'
y, sr = librosa.load(file_path)

# Function to apply pitch shifting
def apply_pitch_shift(audio, n_steps):
    return librosa.effects.pitch_shift(audio, sr=sr, n_steps=n_steps)

# Function to apply trimming
def apply_trim(audio, top_db):
    return librosa.effects.trim(audio, top_db=top_db)[0]

# Function to apply high-pass filtering
def apply_high_pass(audio, cutoff_freq):
    def butter_highpass(cutoff, fs, order=5):
        nyquist = 0.5 * fs
        normal_cutoff = cutoff / nyquist
        b, a = butter(order, normal_cutoff, btype='high', analog=False)
        return b, a

    def butter_highpass_filter(data, cutoff, fs, order=5):
        b, a = butter_highpass(cutoff, fs, order=order)
        y = lfilter(b, a, data)
        return y

    return butter_highpass_filter(audio, cutoff_freq, sr)

# Function to apply low-pass filtering
def apply_low_pass(audio, cutoff_freq):
    def butter_lowpass(cutoff, fs, order=5):
        nyquist = 0.5 * fs
        normal_cutoff = cutoff / nyquist
        b, a = butter(order, normal_cutoff, btype='low', analog=False)
        return b, a

    def butter_lowpass_filter(data, cutoff, fs, order=5):
        b, a = butter_lowpass(cutoff, fs, order=order)
        y = lfilter(b, a, data)
        return y

    return butter_lowpass_filter(audio, cutoff_freq, sr)

# Function to apply modulation
def apply_modulation(audio, depth):
    modulator = np.sin(2 * np.pi * np.arange(len(audio)) / float(sr) * depth)
    return audio * modulator

# Function to apply amplitude scaling
def apply_amplitude_scaling(audio, scaling_factor):
    return audio * scaling_factor

# Function to apply speed change
def apply_speed_change(audio, speed_factor):
    return librosa.effects.time_stretch(audio,rate= speed_factor)

# Function to apply time stretching
def apply_time_stretch(audio, rate):
    return librosa.effects.time_stretch(audio, rate=rate)

# Function to apply randomization
def apply_randomization(audio, factor_range):
    random_factor = np.random.uniform(factor_range[0], factor_range[1])
    return audio * random_factor

# Apply different combinations of augmentations
augmented_audio = apply_speed_change(
    apply_time_stretch(
        apply_modulation(
            apply_amplitude_scaling(
                apply_low_pass(
                    apply_high_pass(
                        apply_trim(
                            apply_pitch_shift(y, n_steps=2),
                            top_db=15
                        ),
                        cutoff_freq=5000
                    ),
                    cutoff_freq=100
                ),
                scaling_factor=1.1
            ),
            depth=0.02
        ),
        rate=0.9
    ),
    speed_factor=1.2
)


# Save and visualize the augmented audio
output_path = '/content/output/augmented_audio.wav'
wavio.write(output_path, augmented_audio, sr, sampwidth=3)

# Plot the original and augmented audio for comparison
plt.figure(figsize=(12, 4))
plt.subplot(2, 1, 1)
plt.plot(np.arange(len(y)) / sr, y)
plt.title('Original Audio')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

plt.subplot(2, 1, 2)
plt.plot(np.arange(len(augmented_audio)) / sr, augmented_audio)
plt.title('Augmented Audio')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()
```
